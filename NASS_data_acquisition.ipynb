```python
# only interested in a few of these columns for this exercise
relevant_farm_codes = ['y17_M059_classRange', 'y17_M061_classRange', 
                       'y17_M063_classRange', 'y17_M066_classRange',
                       'y17_M068_classRange', 'y17_M070_classRange',
                       'y17_M078_classRange', 'y17_M079_classRange',
                       'y17_M080_classRange', 'y17_M081_classRange',
                       'y17_M082_classRange', 'y17_M083_classRange']

boolean_series = NASS_dict.MapID.isin(relevant_farm_codes)
filtered_NASS_dict = NASS_dict[boolean_series]
filtered_NASS_dict.head(15)
```

```python
# can remove irrelevant headers from NASS_farm
final_NASS_farm_columns = ['FIPS', 'y17_M059_classRange', 'y17_M061_classRange', 
                       'y17_M063_classRange', 'y17_M066_classRange',
                       'y17_M068_classRange', 'y17_M070_classRange',
                       'y17_M078_classRange', 'y17_M079_classRange',
                       'y17_M080_classRange', 'y17_M081_classRange',
                       'y17_M082_classRange', 'y17_M083_classRange']

NASS_farm_filtered = NASS_farm.drop(columns=[col for col in NASS_farm if col not in final_NASS_farm_columns])
#NASS_farm_filtered = NASS_farm_filtered.rename(columns = lambda x : str(x)[:-13])
NASS_farm_filtered = NASS_farm_filtered.rename(columns={NASS_farm_filtered.columns[0]: 'FIPS'})
NASS_farm_filtered.head()
```

<h2> Output 1: NASS Farm data </h2>
<p> This is data to capture numeric values with experimental output for variables pertaining to agricultural chemicals used as well as land in farms and land use. The column headers are encoded versions. Lookup table is a separate output. The rows are unique by FIPS, which is a geographic identifier for county/state.<p>

```python
NASS_farm_filtered.to_csv('processed_input_data/NASS_farm.csv', index=False)
```

<h2> Output 2: FIPS Lookup Table </h2>
<p> This is a lookup table of FIPS information. It contains FIPS, county, state, country. This will be used to transform the final dataset, and will serve as a dimension table when NASS_farm data is merged with the eBird data. <p>

```python
NASS_counties.to_csv('processed_input_data/FIPS_lookup.csv', index=False)
```

<h2> Output 3: NASS Header Lookup </h2>
<p> This is a lookup table of data provided by NASS. It describes the codes used to define the column headers in the NASS_farm table. It contains the mapping category and descriptions to keep the headers human readable. The table has been filtered to only include relevant column data. <p>

```python
filtered_NASS_dict.to_csv('processed_input_data/NASS_header_lookup.csv', index=False)
```
